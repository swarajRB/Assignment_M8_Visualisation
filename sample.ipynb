```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper

output_notebook()


# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
# mapper
mapper = LinearColorMapper(palette="Magma256", low=data_heatmap.min(), high=data_heatmap.max())

# Bokeh heatmap
p = figure(x_range=(0, 1), y_range=(0, 1), toolbar_location=None)
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=mapper)
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

# Generate data and histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure for the histogram
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Create the quad glyph to plot the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="skyblue", line_color="white")
       

# Display the plot
show(p)
```
