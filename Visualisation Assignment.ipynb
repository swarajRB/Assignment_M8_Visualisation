## Q.4 4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(sqrt(x**2 + y**2))
x = np.linspace(-5, 5, 100)

 y = np.linspace(-5, 5, 100)

 x, y = np.meshgrid(x, y)

 z = np.sin(np.sqrt(x**2 + y**2))

 'Z': z.flatten()
 
 }
 
 df = pd.DataFrame(data)

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px

# given data 
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

# Converting data into dataframes
df = pd.DataFrame(data)

# Creating 3D plot using plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Adding the title
fig.update_layout(title="3D Surface Plot of z = sin(sqrt(x**2 + y**2))")
            
# Show the plot
fig.show()
```

##  Q.1 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()


# Generate x and y values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

#Plotting the sine wave
p.line(x,y, line_width=2, color='red', legend_label='sin(x)')
show(p)
```

## Q .3 Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a figure for the bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts')

# Bar chart
p.vbar(x=fruits, top=counts, width=0.5 ,
        fill_color=factor_cmap('x', palette=['red', 'orange', 'yellow', 'green'],  factors=fruits))
       
show(p)
```

## Q.4 Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

# Generate data and histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure for the histogram
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Create the quad glyph to plot the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="skyblue", line_color="white")
       

# Display the plot
show(p)
```

## Q.5 Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

 x = np.linspace(0, 1, 10)

 y = np.linspace(0, 1, 10)
 
 xx, yy = np.meshgrid(x, y)


```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper

output_notebook()


# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
# mapper
mapper = LinearColorMapper(palette="Magma256", low=data_heatmap.min(), high=data_heatmap.max())

# Bokeh heatmap
p = figure(x_range=(0, 1), y_range=(0, 1), toolbar_location=None)
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=mapper)
show(p)
```
